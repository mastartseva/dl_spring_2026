## Задание 1. Базовые тензоры и автоград

**Цель:** почувствовать разницу NumPy / PyTorch и понять `requires_grad`, `.backward()`.

**Теория (идея)**  
В PyTorch тензор с `requires_grad=True` участвует в построении вычислительного графа, и по нему можно посчитать градиент скаляра `loss` через `loss.backward()`.  
Градиенты складываются в поле `.grad`. Перед новым шагом их надо обнулять.

**Формула**

Пусть

f(x) = Σ (x_ij^2 + 3 * x_ij) по всем i, j.

Тогда градиент:

∂f/∂x_ij = 2 * x_ij + 3.

**Задание**

1. Создай тензор `x` размера `[3, 4]` со случайными значениями и `requires_grad=True`.
2. Определи скалярную функцию `f(x)` как сумму по всем элементам (структура: `x**2 + 3*x`, затем `.sum()`).
3. Вызови `f.backward()` и выведи `x.grad`.
4. В комментарии выведи аналитический градиент (формула выше) и проверь, что он совпадает с PyTorch через `torch.allclose`.



````python
import torch.nn as nn
import torch
````

````python
x = torch.rand(3, 4, requires_grad = True)
f = (x**2 + 3*x).sum()
````

````python
f.backward()
print(x.grad)
````

````python
x_grad = 2*x + 3
print(x_grad)
````

````python
torch.allclose(x.grad, x_grad)
````

## Задание 2. Свой линейный слой (без `nn.Linear`)

**Цель:** понять, что делает `nn.Linear` и как работает chain rule.

**Теория**  
Линейный слой реализует аффинное преобразование:

y = x * W^T + b,

где  
- x имеет форму `[batch_size, in_features]`,  
- W имеет форму `[out_features, in_features]`,  
- b имеет форму `[out_features]`,  
- результат y имеет форму `[batch_size, out_features]`.

Если W и b созданы с `requires_grad=True`, то после вызова `loss.backward()` PyTorch автоматически заполнит `W.grad` и `b.grad` согласно правилу цепочки (chain rule).

**Задание**

1. Инициализируй данные:

   ```python
   batch_size, in_features, out_features = 5, 3, 2
   x = torch.randn(batch_size, in_features, requires_grad=False)
   W = torch.randn(out_features, in_features, requires_grad=True)
   b = torch.randn(out_features, requires_grad=True)
2. Реализуй «ручной» линейный слой (без использования nn.Linear):

python
y = x @ W.t() + b      # форма y: [batch_size, out_features]

3. Определи функцию потерь:

python
loss = y.pow(2).mean()
loss.backward()

4. Выведи градиенты и прокомментируй:

python
print("W.grad:", W.grad)
print("b.grad:", b.grad)

5. В текстовом комментарии объясни, что:

W.grad — это производная loss по каждому элементу W, усреднённая по всем объектам батча,

b.grad — это производная loss по каждому элементу b, также усреднённая по батчу.

````python
batch_size, in_features, out_features = 5, 3, 2
x = torch.randn(batch_size, in_features, requires_grad = True)
W = torch.randn(out_features, in_features, requires_grad = True)
b = torch.randn(out_features, requires_grad = True)
````

````python
y = x @ W.t() + b
print(y)
````

````python
loss = y.pow(2).mean() #считаем средний лосс по всем объектам батча тк нам нужно одно значение
loss.backward()
print(loss)
````

````python
print("W.grad:", W.grad) # считаем градиенты, т е берем частную производную по каждому элементу относительно лосса
print("b.grad:", b.grad)
````

Задание 3. Мини‑MLP для make_moons
Цель: собрать простую MLP‑модель и руками написать тренировочный цикл.

Теория
Многослойный перцептрон (MLP) — это последовательность слоёв вида:

Linear → нелинейность → Linear → …

Для бинарной классификации удобно иметь один выходной логит и использовать функцию потерь BCEWithLogitsLoss.
Логит — это сырое выходное значение до сигмоиды; функция BCEWithLogitsLoss внутри сама применяет сигмоиду и считает бинарную кросс‑энтропию.

1. Сгенерировать данные
2. Определить модель
3. Написать тренировочный цикл без DataLoader
4. В комментарии поясни:

    почему у BCEWithLogitsLoss на вход подаются сырые логиты,

    как мы считаем accuracy через сравнение logits > 0 с целями y.
    

````python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.2, random_state=0)
x = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).unsqueeze(1)
````

````python
class SimpleMLP(nn.Module):
  def __init__(self):
    super().__init__()
    self.input_layer = nn.Linear(2,16)
    self.hidden_layer = nn.Linear(16,16)
    self.output_layer = nn.Linear(16,1)
    self.act = nn.ReLU()

  def forward(self, x):
    x = self.act(self.input_layer(x))
    x = self.act(self.hidden_layer(x))
    x = self.output_layer(x)

    return x
````

````python
model = SimpleMLP()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1)
criterion = nn.BCEWithLogitsLoss()

for step in range(1000):
  logits = model(x)
  loss = criterion(logits, y)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if step % 100 == 0:
    with torch.no_grad():
      preds = (logits > 0).float()
      acc = (preds == y).float().mean().item()
      print(f"step {step}: loss={loss.item(): .4f}, acc={acc: .3f}")
````
